```python
dirty_df = pd.read_csv("./gottoclean - gottoclean.csv")
```

```python
def create_gold_label(sdg_entry):
    # Initialize an array of zeros with size 17
    gold_label = [0] * 17
    if sdg_entry:
        # Split the entry by commas and convert to integers
        numbers = [int(x) for x in sdg_entry.split(',')]
        # Place the numbers in their respective positions in the array
        for num in numbers:
            if 1 <= num <= 17:
                gold_label[num - 1] = 1
    return gold_label

# Apply the function to create the "Gold Label" column
dirty_df['Gold Label'] = dirty_df['SDG'].apply(create_gold_label)
```

```python
dirty_df = dirty_df.rename(columns={"Rcl Course":"Course", "Rca Desc":"Description"})
dirty_df.to_csv("../data/evaluationData/goldLabelsCategorical.csv", index=False)
```
